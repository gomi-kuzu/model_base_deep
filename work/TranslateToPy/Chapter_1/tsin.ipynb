```python
import numpy as np
import jax

import matplotlib.pyplot as plt
import japanize_matplotlib
import math
```

```python
def tsin(x):
    sum = 0
    n = 8
    for i in range(n):
        if i%2 == 0:
            sum += (1/math.factorial(2*i+1))*x**(2*i+1)
        else:
            sum -= (1/math.factorial(2*i+1))*x**(2*i+1)
    return sum
```

```python
d_tsin = jax.grad(tsin)
```

```python
xs = np.linspace(-5,5,100)
```

```python
tsin_outs = []
d_tsin_outs = []
for x in xs:
    tsin_outs.append(tsin(x))
    d_tsin_outs.append(float(d_tsin(x)))
    
```

```python
plt.plot(xs, np.sin(xs), color = "k", marker="x")
plt.plot(xs, tsin_outs, color= "r", marker="o", alpha=0.5)
plt.show()
```

```python
plt.plot(xs, np.cos(xs), color = "k", marker="x")
plt.plot(xs, d_tsin_outs, color= "r", marker="o", alpha=0.5)
plt.show()
```
