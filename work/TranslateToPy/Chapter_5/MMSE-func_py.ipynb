```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
F = lambda x, var: 1/(np.sqrt(2*np.pi*var))*np.exp(-(1/(2*var))*x**2)
def eta(y, var, p):
    numerator = p*F(y, 1+var)
    denomitor = (1-p)*F(y, var) + p*F(y, 1+var)
    return (y/(1+var))*(numerator/denomitor)
```

```python
xrange = np.arange(-3, 3.01, 0.01)
df = pd.DataFrame()
df["y"] = xrange
```

```python
df["$\sigma^2=0.5$"] = eta(xrange, 0.5, 0.1)
df["$\sigma^2=0.1$"] = eta(xrange, 0.1, 0.1)
df["$\sigma^2=0.01$"] = eta(xrange, 0.01, 0.1)
```

```python
df.plot("y", df.columns[1:])
plt.grid()
```
